```python
## empresas en observacion
import pandas as pd

dict_companies= {
    'BTC-USD':'Bitcoin - Dollar',
    'ANSS':'ANSYS',
    'AVID':'Avid Technology inc.',
    'KO':'Coca Cola company',
    'AAPL':'Apple inc.',
    'AMD':'AMD',
    'CSIQ':'Canadian Solar inc.',
    'AMZN':'Amazon inc.',
    'TSLA':'Tesla motors',
    'PFE':'Pfizer',
    'ABNB':'Airbnb inc'
}

## se convierte dictionary en un DataFrame con keys
companies_data = pd.DataFrame(dict_companies.items(), columns=['stock_symbol','name'])
```

```python
def get_data_company_stocks_days(company_stock_symbol,last_days):
    from datetime import datetime, timedelta
    import pandas_datareader as pdr
    import matplotlib.pyplot as plt

    init_date = datetime.now()-timedelta(days=last_days)
    delta_time = (datetime.today() -  init_date).days

    return(pdr.get_data_yahoo(company_stock_symbol, start=init_date))
```

```python
def plot_stock_data(df_data, company_name, company_stock_symbol, plot_kind):
    import cufflinks as cf

    up_color = '#4CC417'
    down_color = '#DC381F'

    qf = cf.QuantFig(df_data,
                     title='{} - {}'.format(company_stock_symbol, company_name),
                     legend='top',
                     kind=plot_kind,
                     up_color=up_color,
                     down_color=down_color
                    )



    qf.add_bollinger_bands(periods=20,boll_std=2,colors=['black','grey'],fill=True)
    #qf.add_rsi(periods=20,color='java')

    qf.add_adx(color='black')
    qf.add_cci(color='black')
    #qf.add_dmi()
    qf.add_ema(color='red')

    #print('VARIACION: {}'.format())

    qf.iplot()
```

```python
def company_iterator(companies_data):
    #itera por cada empresa del dataframe
    for i in range(companies_data['stock_symbol'].count() - 1):
        company_name = companies_data.iloc[i]['name']
        company_stock_symbol = companies_data.iloc[i]['stock_symbol']
        print('{} - {}'.format(company_stock_symbol,company_name))
        last_days = 120
        plot_kind = 'candle'

        df_data = get_data_company_stocks_days(company_stock_symbol,last_days)
        plot_stock_data(df_data, company_name, company_stock_symbol, plot_kind)
```

```python
company_iterator(companies_data)
```
